```python
import gdal
import matplotlib.pyplot as plt
```

```python
# Working with multiple images

path=r"E:\Spatial_Decision\New folder\Flooding\During\During_Flood.tif"

dataset=gdal.Open(path)
# Get the number of bands 
num=dataset.RasterCount
print(f"The number of bands is {num}")
```

**Plot each band**

```python
plt.figure(figsize=(30,30))

for i in range(1,num+1):
    band=dataset.GetRasterBand(i)
    plt.subplot(1,num,i)
    band_array=band.ReadAsArray()
    plt.imshow(band_array,cmap="gist_gray")
plt.show()
```

**Plot true color**

```python
import numpy as np
band=dataset.ReadAsArray()
# Stack first three bands
stack=np.dstack((band[0],band[1],band[2]))

plt.figure(figsize=(10,10))
plt.imshow(stack)
plt.show()
```
