# 2. Python Loop

### For loop

The for loop is where you iterate over a sequence (such as a list, tuple, dictionary, or string) or other object until you reach the last item in the object.

```python
# Create a list 
mlist=[1,2,3,4,5,["Ha",3,2,4,4,4],{"Name":"Ha", 2:4,9:10},[1,2,3,4],[3,4,5,5]]

for value in mlist:
    print(value)
```

**Loop over index and value**

Using `enumerate()`

```python
list1=[1,2,3,4,"Ha","Van","Tam"]

for index, value in enumerate(list1):
    print(f"Index: {index}  Value: {value}")
```

**`break` and `continue` statement**

`break` statement will stop the loop if the condition is true while `continue` statement will skip the current iteration if the condition is met.

```python
#Break statement
# if i > 5, we stop and quit the loop
for i in range(10):
    if i>5:
        break
    else:
        print(i)
```

```python
# Continue statement
for i in range(10):
    if 5<=i<=8:
        continue
    else:
        print(i)
```

**While loop**

The while loop is similar to an if statement: it executes the code inside of it if some condition is true. The difference is that the while loop will continue to execute as long as the condition is true.

```python
count=0

while True:
    print(f"Number: {count}")
    count+=1
    if count<10:
        break
```

```python
# while loop with a break statement 
count=0

while count<10:
    print(count)
    count+=1
    if count==5:
        break
```

```python
# Small game program
import random
count=0
print("Lucky Numbers! 5 numbers will be generated.")
print("If one of them is a '5', you lose!")
while count<5:
    num=random.randint(1,6)
    print(num)
    if num==5:
        print("Sorry, you lose!")
        break
    count+=1
else:
    print("You win!")
```

**15.** Viết chương trình Python để in mẫu chữ cái in hoa ‘R"



```python
# Nhập số tự nhiên từ 3 đến 20 để đạt chữ R đẹp mắt
n=int(input("Nhập số bạn muốn: "))

for row in range(n):
    for col in range(n):
        if ((row==0 and (col!=0 and col!=n-1)) or (row!=0 and col==0) or ((row!=0 and row<n//2) and col==n-1) or 
            (row==n//2 and (col!=0 and col!=n-1)) or (row==col and row>n//2)):
            
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
